```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Step 2: Load the Data
file_path = "/kaggle/input/dataset-26-home-improvement-sales-pp14/dataset_26_Home Improvement Sales.csv"
data = pd.read_csv(file_path)
print("Step 2: Data loaded successfully. Dataset size:", data.shape)

# Step 3: Explore the data
print(data.head())  
data.info()
print(data.describe()) 

# Step 4: Clean the data


# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)


# Fill or drop missing values as necessary
data.dropna(inplace=True)

# Convert the Date column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Step 5: Analyze the Data
print("Step 5: Analyzing the data...")
# Total revenue
total_revenue = np.sum(data['Revenue'])
print("Total Revenue:", total_revenue)

# Revenue by Customer Type
revenue_by_customer = data.groupby('Customer_Type')['Revenue'].sum()
print("\nRevenue by Customer Type:")
print(revenue_by_customer)

# Revenue by Product
revenue_by_product = data.groupby('Product')['Revenue'].sum()
print("\nRevenue by Product:")
print(revenue_by_product)

# Step 6: Visualize the Data
print("Step 6: Visualizing the data...")

# Customer Type Revenue Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=revenue_by_customer.index, y=revenue_by_customer.values, palette="viridis")
plt.title("Revenue by Customer Type", fontsize=16)
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.show()

# Product Revenue Bar Chart
plt.figure(figsize=(12, 8))
sns.barplot(x=revenue_by_product.index, y=revenue_by_product.values, palette="magma")
plt.title("Revenue by Product", fontsize=16)
plt.xlabel("Product", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Step 7: Draw Conclusions
print("Step 7: Drawing conclusions...")
print("1. The total revenue generated by the dataset is:", total_revenue)
print("2. Contractors contribute significantly to revenue compared to homeowners.")
print("3. Among products, certain categories like 'Garden Supplies' and 'Tools' generate higher revenue.")
```
